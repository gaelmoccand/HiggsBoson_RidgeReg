```python
%matplotlib inline
import matplotlib.image as mpimg
import numpy as np
import matplotlib.pyplot as plt
import os,sys
from PIL import Image
from helper import *
```

```python
# Loaded a set of images
root_dir = "training/"

image_dir = root_dir + "images/"
files = os.listdir(image_dir)
n = 100#len(files) 
print("Loading " + str(n) + " images")
imgs = [load_image(image_dir + files[i]) for i in range(n)]
print(files[0])

gt_dir = root_dir + "groundtruth/"
print("Loading " + str(n) + " images")
gt_imgs = [load_image(gt_dir + files[i]) for i in range(n)]
print(files[0]) 

print(gt_dir)
```

```python
from skimage.transform import rotate
import numpy as np

rot_ang = 5*np.arange(1,71)

for a in range(rot_ang.shape[0]):
    angle = rot_ang[a]
    for i in range(n):
        id = n*(a+1) + i + 1
        # rotate base image
        img_rot = rotate(imgs[i],angle,mode='reflect',preserve_range=True,order=5)
        imageid = "satImage_%.3d" % id
        image_name = image_dir + imageid + ".png"
        print("save image " + image_name)
        Image.fromarray(img_float_to_uint8(img_rot)).save(image_name)
        # rotate gt
        img_rot = rotate(gt_imgs[i],angle,mode='reflect',preserve_range=True,order=5)
        image_name = gt_dir + imageid + ".png"
        print("save image " + image_name)
        Image.fromarray(img_float_to_uint8(img_rot)).save(image_name)
```
